```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
plt.style.use('seaborn-whitegrid')
```

```python
fig = plt.figure()
ax = plt.axes()
```

```python
x = np.linspace(0, 10, 1000)
ax.plot(x, np.sin(x))
```

```python
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))
```

```python
plt.plot(x, np.sin(x - 0), color='blue')   # 
plt.plot(x, np.sin(x - 1), color='g')    # 
plt.plot(x, np.sin(x - 2), color='0.75')   #  0~1 
plt.plot(x, np.sin(x - 3), color='#FFDD44')  # RRGGBB, 00~FF
plt.plot(x, np.sin(x - 4), color=(1.0, 0.2, 0.3))   # RGB 0~1
plt.plot(x, np.sin(x - 5), color='chartreuse')    # HTML
```

```python
plt.plot(x, x + 0, linestyle='solid')  # 
plt.plot(x, x + 1, linestyle='dashed')   # 
plt.plot(x, x + 2, linestyle='dashdot')  # 
plt.plot(x, x + 3, linestyle='dotted')  # 
```

```python
plt.plot(x, x + 0, linestyle='-')  # 
plt.plot(x, x + 1, linestyle='--')   # 
plt.plot(x, x + 2, linestyle='-.')  # 
plt.plot(x, x + 3, linestyle=':')  # 
```

```python
plt.plot(x, x + 1, '-g')  # 
plt.plot(x, x + 2, '--c') # 
plt.plot(x, x + 3, '-.k') # 
plt.plot(x, x + 4, ':r')  # 
```

```python
plt.plot(x, np.sin(x))
plt.xlim(-1, 11)
plt.ylim(-1.5, 1.5)
```

```python
plt.plot(x, np.sin(x))
plt.xlim(12, 1)
plt.ylim(1.5, -1.5)
```

```python
plt.plot(x, np.sin(x))
plt.axis([-1, 11, -1.5, 1.5])
```

```python
plt.plot(x, np.sin(x))
plt.axis('tight')
```

```python
plt.plot(x, np.sin(x))
plt.axis('equal')
```

```python
plt.plot(x, np.sin(x))
plt.title(' SSSD ')
plt.xlabel('x')
plt.ylabel('sin(x)')
```

```python
plt.plot(x, np.sin(x), '-g', label='sin(x)')
plt.plot(x, np.cos(x), ':b', label='cos(x)')
plt.axis('equal')
plt.legend()
```
