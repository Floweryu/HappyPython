```python
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
plt.style.use('classic')
```

```python
import numpy as np
```

```python
x = np.linspace(0, 10, 100)
```

```python
plt.plot(x, np.sin(x))
```

```python
plt.figure()  # 创建图形
```

```python
# 创建两个子图中的第一个
plt.subplot(2, 1, 1) # (行、列、子图编号)
plt.plot(x, np.sin(x))
# 创建两个子图中的第二个
plt.subplot(2, 1, 2)
plt.plot(x, np.cos(x))
```

```python
# 先创建图形网格
# ax 是一个包含两个 Axes 对象的数据
fig, ax = plt.subplots(2)

# 在每一个对象上调用plot()方法
ax[0].plot(x, np.sin(x))
ax[1].plot(x, np.cos(x))
```
